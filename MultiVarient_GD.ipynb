```python
X = df.drop(columns=8)
y = df.iloc[:, 7]
```

```python
X.columns
```

```python
for i in range(1, len(X.columns)+1):
    X[i] = X[i]/np.max(X[i])
X.head()
```

```python
theta = np.array([0]*len(X.columns))
```

```python
m = len(df)
```

```python
def hypothesis(theta, X):
    return theta*X
```

```python
def Cost(X, y, theta):
    y1 = hypothesis(theta, X)
    y1=np.sum(y1, axis=1)
    return sum(np.sqrt((y1-y)**2))/(2*len(X))
```

```python
'''Gradient descent function that compute and update the theta/weight'''
def GradientDescent(X, y, theta, alpha, i):
    J = []  
    k = 0
    while k < i:        
        y1 = hypothesis(theta, X)
        y1 = np.sum(y1, axis=1)
        for c in range(0, len(X.columns)):
            theta[c] = theta[c] - alpha*(sum((y1-y)*X.iloc[:,c])/len(X))
        j = Cost(X, y, theta)
        J.append(j)
        k += 1
    return J, j, theta
```

```python
J, j, theta = GradientDescent(X, y, theta, 0.079, 10000)
```

```python
y_hat = hypothesis(theta, X)
y_hat = np.sum(y_hat, axis=1)
y_hat[0],y[0]
```
